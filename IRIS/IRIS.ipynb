## Univariate Data Analysis

```python
df["species"].value_counts().plot(kind = 'pie',autopct = "%.2f")
plt.show()
```

```python
plt.hist(df['sepal_length'],bins = 20)
plt.show()
```

```python
plt.hist(df['petal_length'],bins = 20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['sepal_width'],color = 'Red',linewidth = 2 )
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['petal_width'],color = 'purple',linewidth = 2)
plt.grid(True)
plt.show()
```

## Bivariative Data Analysis

```python
sns.scatterplot( y = df['petal_length'] , x = df['petal_width'] )
plt.plot([min(df['petal_width']), max(df['petal_width'])], [min(df["petal_length"]), max( df["petal_length"])], linestyle='--', color='red', linewidth=4)
```

```python
sns.scatterplot( x = df['sepal_length'] , y = df['sepal_width'])
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pd.crosstab(df["petal_length"],df['sepal_length']))
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pd.crosstab(df["petal_width"],df['sepal_width']))
plt.show()
```

```python
sns.distplot(df['petal_length'],hist = False)
sns.distplot(df['petal_width'],hist = False)
```

```python
sns.distplot(df['sepal_length'],hist = False)
sns.distplot(df['sepal_width'],hist = False)
```
