## Univariate Data Analysis

```python
# Tv distribution
plt.figure(figsize=(8,5))
sns.distplot(df['TV'],bins = 10)
plt.show()
```

```python
# Tv distribution
plt.figure(figsize=(8,5))
sns.distplot(df['Radio'],bins = 10,hist = False)
plt.show()
```

```python
# Newspaper distribution
plt.figure(figsize=(8,5))
sns.distplot(df['Newspaper'],bins = 10)
plt.show()
```

## Bivariate & Multivariate 

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x = df["TV"],y = df["Sales"],color = "green")
sns.scatterplot(x = df["Radio"],y = df["Sales"],color = "yellow",edgecolor = "black")
sns.scatterplot(x = df["Newspaper"],y = df["Sales"],marker = "s",color = "Red",edgecolor = "white")
plt.plot([min(df['TV']), max(df['TV'])], [min(df["Sales"]), max( df["Sales"])], linestyle='--', color='blue', linewidth=4)
plt.plot([min(df['Radio']), max(df['Radio'])], [min(df["Sales"]), max( df["Sales"])], linestyle='--', color='blue', linewidth=4)
plt.plot([min(df['Newspaper']), max(df['Newspaper'])], [min(df["Sales"]), max( df["Sales"])], linestyle='--', color='blue', linewidth=4)
plt.xlabel("TV vs Newspaper Vs Radio")
plt.show()
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(20,12))
ax = plt.axes(projection = "3d")
fg = ax.scatter3D(df["TV"],df["Radio"],df["Sales"],c = df["TV"],s = 100)
ax.set_xlabel("TV")
ax.set_ylabel("Radio")
ax.set_zlabel("Sales")
plt.colorbar(fg)
```

```python
plt.figure(figsize=(20,12))
ax = plt.axes(projection = "3d")
fg = ax.scatter3D(df["TV"],df["Newspaper"],df["Sales"],c = df["TV"],s = 100)
ax.set_xlabel("TV")
ax.set_ylabel("Newspaper")
ax.set_zlabel("Sales")
plt.colorbar(fg)
```
